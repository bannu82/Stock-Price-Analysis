```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# List of popular Indian stock tickers
tickers = [
    "NHPC.NS", "RELIANCE.NS", "HDFCBANK.NS", "INFY.NS", "TCS.NS", "ICICIBANK.NS",
    "SBIN.NS", "HINDUNILVR.NS", "ITC.NS", "KOTAKBANK.NS", "LT.NS", "AXISBANK.NS",
    "BAJFINANCE.NS", "BHARTIARTL.NS", "MARUTI.NS", "ASIANPAINT.NS", "M&M.NS",
    "SUNPHARMA.NS", "WIPRO.NS", "HCLTECH.NS", "ULTRACEMCO.NS", "TATAMOTORS.NS",
    "TATASTEEL.NS", "TECHM.NS", "ADANIPORTS.NS", "GRASIM.NS", "POWERGRID.NS",
     "BPCL.NS", "IOC.NS"
]

# Download and store data
dataframes = {}

for ticker in tickers:
    data = yf.download(ticker)
    dataframes[ticker] = pd.DataFrame(data)
    break

# Function to plot stock data
def plot_stock_data(ticker, df):
    plt.figure(figsize=(14, 10))

    # Plot Closing Price and Moving Average
    plt.subplot(3, 1, 1)
    sns.lineplot(data=df, x=df.index, y='Close', label='Close')
    df['Moving Average'] = df['Close'].rolling(window=20).mean()
    sns.lineplot(data=df, x=df.index, y='Moving Average', label='20-Day MA', color='orange')
    plt.title(f'{ticker} - Closing Price and Moving Average')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    
    # Plot Volume
    plt.subplot(3, 1, 2)
    sns.barplot(data=df, x=df.index, y='Volume', color='skyblue')
    plt.title(f'{ticker} - Trading Volume')
    plt.xlabel('Date')
    plt.ylabel('Volume')

    # Plot Histogram of Returns
    df['Return'] = df['Close'].pct_change() * 100
    plt.subplot(3, 1, 3)
    sns.histplot(df['Return'].dropna(), bins=50, color='green')
    plt.title(f'{ticker} - Histogram of Returns')
    plt.xlabel('Return (%)')
    plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()

# # Plot for each ticker
# for ticker, df in dataframes.items():
#     plot_stock_data(ticker, df)

# Candlestick Chart using Plotly
def plot_candlestick(ticker, df):
    fig = go.Figure(data=[go.Candlestick(x=df.index,
                                         open=df['Open'],
                                         high=df['High'],
                                         low=df['Low'],
                                         close=df['Close'])])
    fig.update_layout(title=f'{ticker} - Candlestick Chart', xaxis_title='Date', yaxis_title='Price')
    fig.show()

# Plot candlestick charts for each ticker
for ticker, df in dataframes.items():
    plot_candlestick(ticker, df)
    break
```
